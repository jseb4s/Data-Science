# Tu primer árbol de decisión con scikit-learn

```python
#Importamos las librerias principales
import pandas as pd
import matplotlib.pyplot as plt
```

Utilizaremos el **dataset Titanic** de Standford: https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/problem12.html

```python
#Cargamos dataset a utilizar
titanic = pd.read_csv('https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv')
```

## Exploración de datos

```python
#Visualizacion del DataFrame
titanic.head(10)
```

```python
#Eliminamos las columnas que no nos interesan
columns_to_drop = ['Name', 'Fare']
titanic.drop(columns_to_drop, axis = 1, inplace= True)
```

```python
#Renombramos las columnas
titanic.columns = ['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch']
titanic.head()
```

```python
#Analizamos el shape del objeto
titanic.shape
```

```python
#Visualizamos los tipos de datos
titanic.dtypes
```

```python
#Cambiamos los tipos de datos
titanic = pd.get_dummies(titanic, columns=['Sex'], drop_first=True)
```

```python
titanic.dtypes
```

```python
titanic.head()
```

```python
#Reemplazar el DataFrame
titanic.rename(columns = {'Sex_male': 'Sex'}, inplace = True)
```

```python
#Ordenar columnas por nombres
titanic = titanic[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch']]
titanic.head()
```

## Entrenamiento de árbol de decisión con scikit-learn

```python
#Separamos en X e y

X = titanic.drop("Survived", axis = 1)
y = titanic.Survived
```

```python
#Importamos las librerias necesarias para la creacion del modelo
from sklearn.model_selection import train_test_split

#30% para test y 70% para train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state= 00000)
```

```python
#Arbol de Decision
from sklearn.tree import DecisionTreeClassifier

#Creacion del modelo
tree = DecisionTreeClassifier(max_depth = 2, random_state = 00000)

# profundidad del arbol: max_depth (si hay pocos datos, se recomienda una profunidad baja)
```

```python
#Entrenamiento
tree.fit(X_train, y_train)
```

## Evaluación del modelo

```python
#Calculo de las predicciones en Train y Test
y_train_pred = tree.predict(X_train)
y_test_pred = tree.predict(X_test)
```

```python
#Calculo de metricas 
from sklearn.metrics import accuracy_score

#Calculo el accuracy en Train
train_accuracy = accuracy_score(y_train, y_train_pred)

#Calculo el accuracy en Test
test_accuracy = accuracy_score(y_test, y_test_pred)

print(f'El accuracy en train es {train_accuracy}')
print(f'El accuracy en test es {test_accuracy}')
```

```python
#Verificamos el feature importances

# Feature importances indica cual fue la importancia de cada variable (feature) en el modelo a la hora de definir la clasificacion (sobrevive o no sobrevive).
import seaborn as sns

importances = tree.feature_importances_
columns = X.columns
data = pd.DataFrame([importances], columns= columns)
sns.barplot(data, palette = 'bright', saturation = 2.0, edgecolor = 'black', linewidth = 2)
plt.title('Importancia de cada feature')
plt.show()

# Se observa que la variable de sex fue la que tuvo mayor importancia en los resultados de clasificación, seguido de pclass y edad.
```
